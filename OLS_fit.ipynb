```python
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
X = np.array([[1], [1], [2], [2]])
y = 1 * X + 2 * X + 3
# y = np.dot(X, np.array([1, 2])) + 3
reg = LinearRegression(fit_intercept=0).fit(X, y)
reg.score(X, y)

reg.coef_

reg.intercept_

reg.predict(np.array([[3]]))


##############################333333
# Fit the linear regression model
# X = sm.add_constant(X) # Add a constant term to the X variables
model = sm.OLS(y, X).fit()

# Print the model summary to see the p-values
print(model.summary())
```

```python
import pandas as pd
results=model.summary().as_csv()

print(model.params)
print(model.pvalues)
df = pd.read_html(model.summary().tables[1].as_html(),header=0,index_col=0)[0]
df
```

```python
import pandas as pd
import statsmodels.api as sm

# Load the data
data = pd.read_csv("data.csv")

# Create the X and y variables
X = data[['x1', 'x2', 'x3']]
y = data['y']

# Fit the linear regression model
X = sm.add_constant(X) # Add a constant term to the X variables
model = sm.OLS(y, X).fit()

# Print the model summary to see the p-values
print(model.summary())
```
